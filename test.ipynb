```python
import json, os, cv2
import matplotlib.pyplot as plt

with open('CLEVR_v1.0/scenes/CLEVR_train_scenes.json', 'r') as file:
    scenes = json.load(file)['scenes']
with open('CLEVR_v1.0/questions/CLEVR_train_questions.json', 'r') as file:
    questions  = json.load(file)['questions']
```

```python
img_paths = sorted(os.listdir('CLEVR_v1.0/images/train'))
```

```python
imgs = [cv2.imread(f'CLEVR_v1.0/images/train/{path}') for path in img_paths[:10]]
```

```python
img = cv2.imread(f'CLEVR_v1.0/images/train/{img_paths[0]}')
mask = img.copy()
x, y, z = scenes[0]['objects'][1]['pixel_coords']
print(scenes[0]['objects'][1])
z = int(z*2)
mask[(y-z):(y+z), (x-z):(x+z), :] = 0
plt.imshow(mask)
```
